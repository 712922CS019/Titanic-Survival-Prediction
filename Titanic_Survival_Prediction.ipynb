```python
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyRegressor
from  sklearn.metrics import mean_squared_error,mean_absolute_error
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
# Assuming 'df' refers to the Titanic dataset loaded previously.

# Define target variable 's'
s = df["Survived"]

# Define feature DataFrame 'o'
o = df.drop(columns=["Survived"], axis=1)

# Preprocess features: handle missing values and encode categorical columns
X_processed = o.copy() # copy is used to saty same dataset even after filtering this is must for preproceesing

# Drop columns that are typically not useful or contain too many unique values for simple encoding(NO RELATION WITH THE TARGET VARIABLE)
X_processed = X_processed.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin']) # this only supports column not axis=1 or 0

# Handle missing 'Age' values by filling with the median
if X_processed['Age'].isnull().any():
    X_processed['Age'].fillna(X_processed['Age'].median(), inplace=True)

# Handle missing 'Embarked' values by filling with the most frequent value
if X_processed['Embarked'].isnull().any():
    X_processed['Embarked'].fillna(X_processed['Embarked'].mode()[0], inplace=True)

# One-hot encode categorical features 'Sex' and 'Embarked'
X_processed = pd.get_dummies(X_processed, columns=['Sex', 'Embarked'], drop_first=True)

# Now perform the train-test split with the correctly prepared features (X_processed) and target (s)
train_x, test_x, train_y, test_y = train_test_split(X_processed, s, test_size=0.2, random_state=42)

gokul=RandomForestClassifier(
    n_estimators=300, # depth it goes till 300,
    max_depth=5,# for each tree it goes depth into 5,
    min_samples_split=5,# the split of random 5,
    min_samples_leaf=3,# inside the split of one by one it goes leaf children like 3
    random_state=42
)
gokul2=gokul.fit(train_x,train_y)
predictio=gokul2.predict(test_x)
print("train",gokul2.score(train_x,train_y))
print("test",gokul2.score(test_x,test_y))
```

```python
from sklearn.metrics import accuracy_score,classification_report # we use this because dataset is classifier
gokull=accuracy_score(test_y,predictio) # test_y and prediction cobines and gives the accuracy
gokull2=classification_report(test_y,predictio)# same combining
print(gokull)
print(gokull2)
```

```python
from sklearn.model_selection import cross_val_score
gokulam=cross_val_score(gokul,X_processed,s,cv=5,scoring="accuracy")
print(gokulam)
```
